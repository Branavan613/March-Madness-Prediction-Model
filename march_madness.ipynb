```python
def prev_finder(row, dd, selector):
    year = row["YEAR"]
    team = row["TEAM ID"]
    if year == 2021:
        year -= 1
    if not dd[dd["YEAR"]==(year-1)].empty:
        val = dd[(dd["YEAR"]==(year-1)) & (dd["TEAM ID"] == team)][selector]
        if not val.empty:
            return dd[(dd["YEAR"]==(year-1)) & (dd["TEAM ID"] == team)][selector].iloc[0]
        else:
            if selector == "ROUND":
                return 128
            else:
                return 17
    else:
        return "NaN"


ddd=dd.copy()
ddd=ddd.drop_duplicates()
```

```python
ddd["PREV SEED"] = ddd.apply(lambda row: prev_finder(row, ddd, "SEED"), axis=1)
ddd["PREV ROUND"] = ddd.apply(lambda row: prev_finder(row, ddd, "ROUND"), axis=1)
```

```python
def scorer1(comb):
    comb_u = comb[comb["ROUND"] == round(comb["PREDICTIONS"])]
    s = 0
    for pred in comb_u["ROUND"]:
        val = pred
        if pred != 1:
            while val > 0:
                s += 2**(val-1)
                val -= 1            
    return s

def scorer2(comb):
    comb_u = comb[comb["ROUND"] <= round(comb["PREDICTIONS"])]
    s = 0
    for pred in comb_u["ROUND"]:
        val = pred
        if pred == 1:
            s += 1
        else:
            while val > 0:
                s += 2**(val-1)
                val -= 1
    return s/2

def diff(comb):
    comb["DIFF"] = abs(comb["ROUND"] - round(comb["PREDICTIONS"]))
    return comb["DIFF"].sum()/64

def corr(comb):
    return len(comb[comb["ROUND"] == round(comb["PREDICTIONS"])])/64
def maxdiff(comb):
    comb["DIFF"] = abs(comb["ROUND"] - round(comb["PREDICTIONS"]))
    return max(comb["DIFF"])
diff(comb)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=50, random_state=1, min_samples_split=10)
```

```python
def backtest(data, model, predictors):
    aps = []
    score = []
    all_predictions = []
    diffs = []
    farthest = []
    correct = []
    for year in years[2:]:
        print(year)
        train = data[(data["YEAR"] < year)]
        test = data[data["YEAR"] == year]
        model.fit(train[predictors], train["ROUND"])
        prediction = model.predict(test[predictors])
        prediction = pd.DataFrame(prediction, columns=['PREDICTIONS'], index =test.index)
        comb = pd.concat([test[["TEAM", "ROUND", "YEAR"]], prediction], axis=1)
        comb = comb.drop_duplicates(subset="TEAM", keep= "first")
        score.append(scorer1(comb))
        diffs.append(diff(comb))
        farthest.append(maxdiff(comb))
        all_predictions.append(comb)
        correct.append(corr(comb))
        aps.append(mean_squared_error(comb["ROUND"], comb["PREDICTIONS"]))
        
    error_ext = {
        "meansquared" : [(sum(aps)/len(aps)), aps],
        "scores" : [(sum(score)/len(score)), score],
        "differences" : [(sum(diffs)/len(diffs)), diffs],
        "correct predictions" : [(sum(correct)/len(correct)), correct],
        "worst prediction" : farthest
    }
    error_min = [(sum(correct)/len(correct)), (sum(aps)/len(aps)), (sum(score)/len(score)), (sum(diffs)/len(diffs)), sum(farthest)]
    
    return all_predictions, error_ext, error_min
```

```python
predicts, error, error_min = backtest(clean_d, rf, predictors)
```
